# **Probabilidades del tablero de colores**

Supongamos que tenemos un tablero con cuatro sectores. Lanzamos dos monedas y avanzamos en sentido horario, del número de caras que aparezcan. Si repetimos el procedimiento. ¿Estarán todos los sectores igual de probables cuando pase mucho tiempo?


Donde tomamos:

$x_n = $ la probabilidad de estar en el sector naranja tras $n$ lanzamientos.

$y_n = $ la probabilidad de estar en el sector azul tras $n$ lanzamientos.

$z_n = $ la probabilidad de estar en el sector verde tras $n$ lanzamientos.

$t_n = $ la probabilidad de estar en el sector rojo tras $n$ lanzamientos.



$P\{Naranja|Naranja\} = 1/4$

$P\{Naranja|Azul\} = 0$

$P\{Naranja|Verde\} = 1/4$

$P\{Naranja|Rojo\} = 1/2$

$P\{Azul|Azul\} = 1/4$

$P\{Azul|Verde\} = 0$

$P\{Azul|Rojo\} = 1/4$

$P\{Azul|Naranja\} = 1/2$

$P\{Verde|Verde\} = 1/4$

$P\{Verde|Rojo\} = 0$

$P\{Verde|Naranja\} = 1/4$

$P\{Verde|Azul\} = 1/2$

$P\{Rojo|Rojo\} = 1/4$

$P\{Rojo|Naranja\} = 0$

$P\{Rojo|Azul\} = 1/4$

$P\{Rojo|Verde\} = 1/2$


En general, matricialmente se ve de la siguiente manera

En cualquier caso se trata de resolver:
$X_n = A \cdot X_{n-1} = A^2 \cdot X_{n-2} = \cdots = A^n \cdot X_0$

Se trata entonces de calcular $A^n$.
Las ecuaciones de Chapman-Kolmogorov proporcionan un método para calcular las probabilidades de transición de n pasos.

Hagamos entonces el código que nos calcule las probabilidades  $x_n$, $y_n$, $z_n$, $t_n$ y las grafique, siguiendo el problema del tablero de colores y tomando $n$ iteraciones (tomaremos $n=50$ en este caso).

```python
import matplotlib.pyplot as plt
```

Condiciones iniciales:

```python
x = 1  # empezamos en el sector x
y = 0
z = 0
t = 0

n = 50  # número de iteraciones
```

Creamos listas para guardar los valores

```python
x_i = []
y_i = []
z_i = []
t_i = []
```

Proceso iterativo:

```python
for i in range(n):
    xi = (1/4)*x + 0*y + (1/4)*z + (2/4)*t
    yi = (2/4)*x + (1/4)*y + 0*z + (1/4)*t
    zi = (1/4)*x + (2/4)*y + (1/4)*z + 0*t
    ti = 0*x + (1/4)*y + (2/4)*z + (1/4)*t

    # Guardar los resultados
    x_i.append(xi)
    y_i.append(yi)
    z_i.append(zi)
    t_i.append(ti)

    x, y, z, t = xi, yi, zi, ti  # actualizamos para la siguiente iteración
```

Graficando la evolución de las probabilidades en el tablero de colores:

```python
plt.figure(figsize=(10,6))
plt.plot(x_i, label='$x_n$')
plt.plot(y_i, label='$y_n$')
plt.plot(z_i, label='$z_n$')
plt.plot(t_i, label='$t_n$')
plt.legend()
plt.grid(True)
plt.show()
```

Calculando el estado estacionario:

```python
print("Distribución final aproximada (n =", n,"):")
print(f"x = {x}")
print(f"y = {y}")
print(f"z = {z}")
print(f"t = {t}")
```

Por lo tanto, cuando pase mucho tiempo (después de n iteraciones), los sectores si son igual de probables.
